```python
from numpy import array_equal
from utils import get_pair, one_hot_decode
from keras.models import Sequential
from keras.layers import LSTM, Dense, TimeDistributed, RepeatVector
from attention_decoder import AttentionDecoder
```

### Creacion del conjunto de datos

```python
n_features = 50
n_timesteps_in = 5
n_timesteps_out = 2

X, y = get_pair(5, 2, 50)
print("'X' and 'y' shapes: \n")
print(X.shape, y.shape)
print('\nX=%s, y=%s' % (one_hot_decode(X[0]), one_hot_decode(y[0])))
print("\nX:")
print(X)
print("\ny:")
print(y)
```

### Modelo RNN-LSTM
Este modelo cuenta con 2 capas y tiene cuenta con 308,750 parametros

```python
model = Sequential()
model.add(LSTM(150, input_shape=(n_timesteps_in, n_features)))
model.add(RepeatVector(n_timesteps_in))
model.add(LSTM(150, return_sequences=True))
model.add(TimeDistributed(Dense(n_features, activation='softmax')))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])
```

Entrenamiendo del modelo con 5000 datos diferentes

```python
for epoch in range(5000):
    X,y = get_pair(n_timesteps_in, n_timesteps_out, n_features)
    model.fit(X, y, epochs=1, verbose=2)
```

Computo del accuracy para el modelo 

```python
total, correct = 100, 0
for _ in range(total):
    X,y = get_pair(n_timesteps_in, n_timesteps_out, n_features)
    yhat = model.predict(X, verbose=0)
    if array_equal(one_hot_decode(y[0]), one_hot_decode(yhat[0])):
        correct += 1
print('Accuracy: %.2f%%' % (float(correct)/float(total)*100.0))
```

### Modelo RNN-LSTM con el mecanismo de atencion
Este modelo tiene 363,900 parametros 53195 parametros mas que la version anterior

```python
model = Sequential()
model.add(LSTM(150, input_shape=(n_timesteps_in, n_features), return_sequences=True))
model.add(AttentionDecoder(150, n_features))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])
```

Se entrena y se evalua el modelo siguiendo la misma metodologia anterior

```python
for epoch in range(5000):
    X,y = get_pair(n_timesteps_in, n_timesteps_out, n_features)
    model.fit(X, y, epochs=1, verbose=2)
```

```python
total, correct = 100, 0
for _ in range(total):
    X,y = get_pair(n_timesteps_in, n_timesteps_out, n_features)
    yhat = model.predict(X, verbose=0)
    if array_equal(one_hot_decode(y[0]), one_hot_decode(yhat[0])):
        correct += 1
print('Accuracy: %.2f%%' % (float(correct)/float(total)*100.0))
```
